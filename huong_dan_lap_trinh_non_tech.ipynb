## 📌 Phần 3: Biến (Variables)

**Biến** là "hộp lưu trữ" thông tin trong máy tính.

```
tên_biến = giá_trị
```

Ví dụ thực tế: Bạn có một chiếc hộp dán nhãn "tuổi" và bỏ số 25 vào trong.

````python
# Tạo biến
ten = "Nguyễn Văn A"
tuoi = 25
chieu_cao = 1.70
da_co_viec_lam = True

# In ra giá trị
print("Tên:", ten)
print("Tuổi:", tuoi)
print("Chiều cao:", chieu_cao, "m")
print("Đã có việc làm:", da_co_viec_lam)
````

### Các kiểu dữ liệu cơ bản

| Kiểu | Ví dụ | Dùng để |
|------|-------|--------|
| `str` (chuỗi) | `"Xin chào"` | Văn bản |
| `int` (số nguyên) | `42` | Đếm, tính toán |
| `float` (số thực) | `3.14` | Đo lường |
| `bool` (logic) | `True / False` | Điều kiện |

````python
# Kiểm tra kiểu dữ liệu
print(type("Xin chào"))   # str
print(type(42))            # int
print(type(3.14))          # float
print(type(True))          # bool
````

## 📌 Phần 4: Câu Lệnh Điều Kiện (If/Else)

Máy tính có thể **ra quyết định** dựa trên điều kiện.

Tương tự như cuộc sống:
> *"NẾU trời mưa THÌ mang ô, NGƯỢC LẠI thì mặc áo mỏng"*

````python
nhiet_do = 35  # độ C

if nhiet_do > 30:
    print("Trời nóng! Uống nhiều nước nhé.")
elif nhiet_do > 20:
    print("Thời tiết dễ chịu.")
else:
    print("Trời lạnh! Mặc áo ấm vào.")
````

````python
# Bài tập: Kiểm tra đủ tuổi bầu cử
tuoi = 20

if tuoi >= 18:
    print("Bạn đủ tuổi bầu cử!")
else:
    print(f"Bạn chưa đủ tuổi. Còn {18 - tuoi} năm nữa.")
````

## 📌 Phần 5: Vòng Lặp (Loops)

Vòng lặp giúp **tự động lặp lại** công việc nhàm chán.

Ví dụ: Thay vì viết `print` 10 lần, ta dùng vòng lặp!

````python
# Vòng lặp for — lặp qua một dãy số
print("Đếm từ 1 đến 5:")
for so in range(1, 6):
    print(so)
````

````python
# Lặp qua danh sách
mon_hoc = ["Toán", "Văn", "Anh", "Lý", "Hóa"]

print("Danh sách môn học:")
for mon in mon_hoc:
    print("-", mon)
````

````python
# Vòng lặp while — lặp khi điều kiện còn đúng
dem = 0
while dem < 5:
    print(f"Lần {dem + 1}: Tôi đang học lập trình!")
    dem = dem + 1
````

## 📌 Phần 6: Hàm (Functions)

**Hàm** = Đóng gói một đoạn code để tái sử dụng.

Giống như công thức nấu ăn — viết một lần, dùng nhiều lần!

````python
# Định nghĩa hàm
def chao_hoi(ten):
    print(f"Xin chào, {ten}! Chào mừng đến với lập trình.")

# Gọi hàm
chao_hoi("An")
chao_hoi("Bình")
chao_hoi("Châu")
````

````python
# Hàm có giá trị trả về
def tinh_bmi(can_nang_kg, chieu_cao_m):
    bmi = can_nang_kg / (chieu_cao_m ** 2)
    return round(bmi, 2)

def danh_gia_bmi(bmi):
    if bmi < 18.5:
        return "Thiếu cân"
    elif bmi < 25:
        return "Bình thường"
    elif bmi < 30:
        return "Thừa cân"
    else:
        return "Béo phì"

# Sử dụng
bmi = tinh_bmi(60, 1.65)
print(f"BMI của bạn: {bmi}")
print(f"Đánh giá: {danh_gia_bmi(bmi)}")
````

## 📌 Phần 7: Danh Sách (Lists)

**List** = Danh sách các phần tử, giống như danh sách mua sắm.

````python
# Tạo và thao tác với list
danh_sach_mua_sam = ["Sữa", "Trứng", "Bánh mì"]

# Thêm phần tử
danh_sach_mua_sam.append("Nước cam")

# Xóa phần tử
danh_sach_mua_sam.remove("Trứng")

# In toàn bộ
print("Danh sách mua sắm:")
for item in danh_sach_mua_sam:
    print(f"  ✓ {item}")

print(f"\nTổng số món: {len(danh_sach_mua_sam)}")
````

## 🎯 Bài Tập Tổng Hợp

Viết chương trình tính điểm trung bình và xếp loại học sinh.

````python
def tinh_diem_trung_binh(diem_cac_mon):
    tong = sum(diem_cac_mon)
    so_mon = len(diem_cac_mon)
    return tong / so_mon

def xep_loai(diem_tb):
    if diem_tb >= 9.0:
        return "Xuất sắc"
    elif diem_tb >= 8.0:
        return "Giỏi"
    elif diem_tb >= 6.5:
        return "Khá"
    elif diem_tb >= 5.0:
        return "Trung bình"
    else:
        return "Yếu"

# Dữ liệu học sinh
hoc_sinh = {
    "An": [8.5, 9.0, 7.5, 8.0, 9.5],
    "Bình": [6.0, 7.0, 5.5, 6.5, 7.0],
    "Châu": [9.5, 9.0, 9.8, 9.2, 9.6]
}

print("=" * 40)
print("KẾT QUẢ HỌC TẬP")
print("=" * 40)

for ten, diem in hoc_sinh.items():
    diem_tb = tinh_diem_trung_binh(diem)
    loai = xep_loai(diem_tb)
    print(f"{ten:10} | ĐTB: {diem_tb:.1f} | Xếp loại: {loai}")
````
